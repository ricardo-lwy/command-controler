```python

########now we start process3
####! input :   1.sousimgfiles
####            2.excelfile
excelname='05122022'
#!SHEETNAME
import pandas as pd
df=pd.read_excel(excelname+'.xlsx',sheet_name='Orders')
#df=pd.read_excel(excelname+'.xlsx',sheet_name='Sheet1')


###库
import cv2
import numpy as np
from datetime import datetime
import pytz
from math import *
##识别图片模型（下载了很久。。。。。）
import easyocr
import os, shutil
import img2pdf
import glob
import fitz
import re
import PyPDF2
from PIL import Image,ImageFont,ImageDraw
import matplotlib.pyplot as plt


###########processe2 finished 
####output : image with name(tran id)

from pandas.core.frame import DataFrame
from openpyxl.writer.excel import ExcelWriter

def mkdir(path):
    folder = os.path.exists(path)
    if not folder:  # 判断是否存在文件夹如果不存在则创建为文件夹
        os.makedirs(path)  # makedirs 创建文件时如果路径不存在会创建这个路径
    else:
        pass









#############################################################
#########################################################

sumline = len(df)
sousimgdir = './sousimgs/'
file_list = []
for items in os.walk(sousimgdir, topdown=False):
    ##遍历后得到的结果
    file_list = items[2]
if(sumline>len(file_list)):
    print("订单数大于邮票数！！！！！！！！！！！！！！！！！！！！！！！！")
else:
    print("预计运行完成后剩余%d张图片可以使用"%(len(file_list)-sumline))







###output ： 1.resultfile.jpg       放入res-image 文件夹
###          2.mv usedfile          将结果文件和源文件一起打包放到finished-order-excel中
###          3.mvsousimg            将用过的图片放进used-sousimg中
####         剩余的sousimg数量
ID=[]
NAME=[]
FIRN=[]
LASN=[]
ADRE=[]
TRANID=[]
usedimgpath="./used-sousimg/"
resimgpath="./res-image/"
```

```python
#############################start process 3
############################################################################################
############################################################################################
mkdir(usedimgpath + excelname )
mkdir(resimgpath + excelname)
index_filelist=0
for i in range(0,sumline):


    nowid=df.values[i][0]
    nowname=df.values[i][1]         #！getname as filename
    #guest info

    nowfirname=str(df.values[i][2]) if str(df.values[i][2])!='nan' else ''
    nowlasname=str(df.values[i][3]) if str(df.values[i][3])!='nan' else ''
    nowadr=str(df.values[i][4]) if str(df.values[i][4])!='nan' else ''
    if(str(df.values[i][5])=='nan'):
        nowadr2=''
    else:
        nowadr2=df.values[i][5]
    
    nowcity=str(df.values[i][6])+' '+df.values[i][7] if len(str(df.values[i][6]))==5 else '0'+str(df.values[i][6])+' '+df.values[i][7]
    ##!france citycode
    if(len(str(df.values[i][6]))!=5):
        df.loc[i,('Shipping: Zip')]='0'+str(df.values[i][6])


    nowcountry=str(df.values[i][8]) if str(df.values[i][8])!='nan' else ''
    ####write
    df.loc[i,('Fulfillment: Tracking Number')]=file_list[index_filelist][:-4]
    df.loc[i,('Fulfillment: Tracking Company')]='La Poste'
    
    


    ###生成图片，然后将原来的图片放到used文件夹中,生成的图片放到res
    ###文件夹中

    imgfullpath=sousimgdir+file_list[index_filelist]
    #拼接图片

    ##timbre
    concatimg=Image.new("RGB", (1000, 1000), (255, 255, 255))
    im2=Image.open(imgfullpath).resize((581,305), Image.Resampling.BILINEAR)
    concatimg.paste(im2,(355,0))
    ##!timbre



    #info
    #! ligne 1
    allinfo=nowfirname+' '+nowlasname+'\n'
    #! ligne 2
    allinfo+=nowadr+'\n'
    #! ligne 3 
    allinfo+=str(nowadr2)+'\n'
    #! ligne 4
    allinfo+=nowcity+'\n'
    allinfo+=nowcountry
    ##length hight and color
    im = Image.new("RGB", (600,260), (255, 255, 255))
    dr = ImageDraw.Draw(im)
    font = ImageFont.truetype(os.path.join("C:/Windows/Fonts/timesbd.TTF"), 28)
    ##第一个参数是前后，第二个参数是上下
    dr.text((0, 0), allinfo, font=font, fill="#000000")
    concatimg.paste(im,(400,340))
    ##!info

    #info preciousse
    infopreciousse="            Expéditeur : "+'\n'
    infopreciousse+=" INFINITUS - PRECIOUSSE"+'\n'
    infopreciousse+="5 BIS RUE DE SAINT-GOBAIN"+'\n'
    infopreciousse+="    CENTRE CIFA LOT3211 "+'\n'
    infopreciousse+="     93300 AUBERVILLIERS "+'\n'
    im3 = Image.new("RGB", (350,260), (255, 255, 255))
    dr2 = ImageDraw.Draw(im3)
    font = ImageFont.truetype(os.path.join("C:/Windows/Fonts/ARIAL.TTF"), 16)
    ##第一个参数是前后，第二个参数是上下
    dr2.text((50, 0), infopreciousse, font=font, fill="#000000")
    #concatimg.paste(im3,(500,400))
    concatimg.paste(im3,(0,80))
    #!info preciousse

    

    ###图片处理！以及转pdf
    box=(0,0,500,1000)
    concatimg=concatimg.rotate(90)
    concatimg=concatimg.crop(box)







    ###拼接图片的保存
    concatimg.save(resimgpath + excelname+'/'+nowname+'.jpg',quality=95)


  


    


    
    
    ##被command excel使用过的图片文件夹
    
    shutil.move(imgfullpath,usedimgpath + excelname)
    
   
    #计数器+1
    index_filelist+=1

 ###excel文件保存
 ######目前这一步需要在里面添加几列.19112022
 #################怕影响前面的代码在这里加一下列好了
df.drop('Fulfillment: Status', axis=1,inplace=True) 
col_name = df.columns.tolist()
col_name.insert(col_name.index('Name')+1,'Command')# 在所选列前面插入另一列，index后+1则是在后面加另一列
col_name.insert(col_name.index('Command')+1,'Line: Type')
col_name.insert(col_name.index('Line: Type')+1,'Fulfillment: ID')
  
#col_name.insert(col_name.index('Fulfillment: ID')+1,'Fulfillment: Status')
#！ tracking number insert
col_name.insert(col_name.index('Fulfillment: ID')+1,'Fulfillment: Status')
col_name.insert(col_name.index('Fulfillment: Tracking Number')+1,'Fulfillment: Tracking URL')
col_name.insert(col_name.index('Fulfillment: Tracking URL')+1,'Fulfillment: Send Receipt')
#!  drop    df.drop(['B', 'C'], axis=1)   axis=1 column

df=df.reindex(columns=col_name)
#df.drop('Fulfillment: Status', axis=1,inplace=True) 
#!toexcel


trackingurllink="https://www.laposte.fr/outils/suivre-vos-envois?code="
for i in range(0,sumline):
    #@command
    df.loc[i,('Command')]=r"UPDATE"
    df.loc[i,('Line: Type')]=r"Fulfillment Line"
    df.loc[i,('Fulfillment: Status')]=r"Success"
    df.loc[i,('Fulfillment: Send Receipt')]=r"TRUE"

    #@ TRACKING URL
    trackingnum=df.values[i][14]        #will be changed by action reindex
    
    fulltrackingnum=str(trackingurllink)+str(trackingnum)
    df.loc[i,('Fulfillment: Tracking URL')]=fulltrackingnum

#finish
    






df.to_excel("%s.xlsx"%(excelname),index=False,sheet_name="Orders")


print("success concat sousimg")
```

```python
#！ to pdf
# #先转单页再拼接
def jpg2pdf(jpgFile): 
    global imglist

    path,fileName = jpgFile.rsplit('\\',1)
    preName,postName = fileName.rsplit('.',1)


    img = Image.open(jpgFile)
    imglist.append(img)
    return img.save(path+"\\"+preName+'.pdf', "PDF", resolution=100.0, save_all=True)
def jpg2pdfByPath(pathName):

    global imglist
  
    imglist = []
    imgfile = ''
    files = os.listdir(pathName)
    for f in files:
        if f.lower().find(".jpg")>0 :

            jpg2pdf(pathName+'\\'+f)
            imgfile = f
         

    imgMerge = imglist.pop(0) #取出第一个图片示例

    imgMerge.save(pathName+r'\merge.pdf',"PDF", resolution=100.0, save_all=True, append_images=imglist)
    print("all images processed!")


jpg2pdfByPath(resimgpath + excelname)
```
